# Exercise 2

```python
#import necessary modules
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
```

```python
# download the Punkt tokenizer and stopwords models for the NLTK library

nltk.download('punkt')
nltk.download('stopwords')
```

```python
#Tokenize the text into sentences using sent_tokenize
sentences = sent_tokenize(text)
```

```python
#Tokenize each sentence into words, then remove stopwords
stop_words = set(stopwords.words('english'))
cleaned_words = []
```

```python
print(stop_words)
```

```python
for sentence in sentences:
    words = word_tokenize(sentence)
    words = [word.lower() for word in words if word.isalnum() and word.lower() not in stop_words]
    cleaned_words.extend(words)
```

```python
#Combine cleaned words into a single string

cleaned_text = ' '.join(cleaned_words)
```

```python
cleaned_text
```

```python
#WordCloud to generate a word cloud image

wordcloud = WordCloud( background_color='white').generate(cleaned_text)
##Downloaded the TrueType font and used the .ttf file path in font_path
```

```python
#Display the word cloud image using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
